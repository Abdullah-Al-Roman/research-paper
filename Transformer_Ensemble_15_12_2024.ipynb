```python
import pandas as pd
import numpy as np
import torch
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer, DataCollatorWithPadding
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
from sklearn.model_selection import train_test_split
```

```python
# Define the models we'll use in our ensemble
model_names = [
    "bangla-bert-base",
    "bert-base-multilingual-uncased",
    "distilbert-base-uncased",
    "xlm-roberta-base"
]
```

```python
# Load and preprocess the datasets
def load_and_preprocess_data(file_path):
    df = pd.read_csv(file_path)
    # Ensure 'text' column exists and contains string data
    if 'text' not in df.columns:
        raise ValueError(f"'text' column not found in {file_path}")
    df['text'] = df['text'].astype(str)
    return df

try:
    train_df = pd.read_csv('/content/Traning.xlsx - Sheet1 (7).csv')
    val_df = pd.read_csv('/content/Val.xlsx - Sheet1 (2).csv')
    test_df = pd.read_csv('/content/Test.xlsx - Sheet1 (2).csv')
    temp_df = pd.read_csv('/content/Val.xlsx - Sheet1 (2).csv')
    val_df['labels'] = temp_df['labels']
except Exception as e:
    print(f"Error loading data: {e}")
    raise
```

```python
# Combine train and validation for stratified split
combined_df = pd.concat([train_df, val_df], ignore_index=True)
```

```python
# Perform stratified split
train_df, val_df = train_test_split(combined_df, test_size=0.1, stratify=combined_df['labels'], random_state=42)
```

```python
# Convert to Hugging Face datasets
train_dataset = Dataset.from_pandas(train_df)
val_dataset = Dataset.from_pandas(val_df)
test_dataset = Dataset.from_pandas(test_df)
```

```python
# Helper function for computing metrics
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='weighted')
    acc = accuracy_score(labels, preds)
    return {
        'accuracy': acc,
        'f1': f1,
        'precision': precision,
        'recall': recall
    }
```

```python
# Function to train a single model
def train_model(model_name):
    tokenizer = AutoTokenizer.from_pretrained(model_name)

    def tokenize_function(examples):
        # Ensure 'text' is a list of strings
        texts = [str(text) for text in examples['text']]
        return tokenizer(texts, padding="max_length", truncation=True, max_length=128)

    tokenized_train = train_dataset.map(tokenize_function, batched=True)
    tokenized_val = val_dataset.map(tokenize_function, batched=True)

    data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

    training_args = TrainingArguments(
        output_dir=f"./results_{model_name.split('/')[-1]}",
        num_train_epochs=5,
        per_device_train_batch_size=16,
        per_device_eval_batch_size=64,
        warmup_steps=500,
        weight_decay=0.01,
        logging_dir=f'./logs_{model_name.split("/")[-1]}',
        logging_steps=10,
        evaluation_strategy="epoch",
        save_strategy="epoch",
        load_best_model_at_end=True,
        metric_for_best_model="f1",
    )

    model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=5)

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=tokenized_train,
        eval_dataset=tokenized_val,
        tokenizer=tokenizer,
        data_collator=data_collator,
        compute_metrics=compute_metrics,
    )

    trainer.train()

    return trainer, tokenizer
```

```python
# Train all models
trained_models = []
for model_name in model_names:
    print(f"Training model: {model_name}")
    try:
        trainer, tokenizer = train_model(model_name)
        trained_models.append((trainer, tokenizer))
    except Exception as e:
        print(f"Error training model {model_name}: {e}")
        continue
```

```python
# Function to get predictions from a single model
def get_predictions(trainer, tokenizer, dataset):
    def tokenize_function(examples):
        # Ensure 'text' is a list of strings
        texts = [str(text) for text in examples['text']]
        return tokenizer(texts, padding="max_length", truncation=True, max_length=128)

    tokenized_dataset = dataset.map(tokenize_function, batched=True)
    predictions = trainer.predict(tokenized_dataset)
    return predictions.predictions
```

```python
import numpy as np
import pandas as pd
import torch

def get_predictions(trainer, tokenizer, dataset):
    """
    Generates predictions using the given trainer and dataset.
    """
    try:
        # Generate predictions using the trainer's predict method
        output = trainer.predict(dataset)
        return output.predictions  # Returns logits or label indices
    except Exception as e:
        print(f"Error in get_predictions: {e}")
        raise

# Initialize variables
all_predictions = []

# Check device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Process each trained model
for trainer, tokenizer in trained_models:
    model_name = trainer.model.name_or_path
    print(f"Processing model: {model_name}")

    try:
        # Move model to the correct device
        trainer.model.to(device)

        # Get predictions for the test dataset
        predictions = get_predictions(trainer, tokenizer, test_dataset)
        print(f"Model {model_name} predictions shape: {predictions.shape}")

        # Convert logits to label indices if necessary
        if predictions.ndim > 1:
            predictions = predictions.argmax(axis=1)

        # Append predictions to the list
        all_predictions.append(predictions)
    except Exception as e:
        print(f"Error with model {model_name}: {e}")
        continue

# Ensemble predictions using majority voting
if all_predictions:
    try:
        # Stack predictions from all models (shape: [num_models, num_samples])
        ensemble_predictions = np.stack(all_predictions)

        # Perform majority voting along the model axis
        majority_votes = np.apply_along_axis(
            lambda x: np.bincount(x).argmax(), axis=0, arr=ensemble_predictions
        )

        # Create a DataFrame with results
        results_df = pd.DataFrame({
            'text': test_df['text'],  # Original text from the test DataFrame
            'label': majority_votes  # Ensemble-predicted labels
        })

        # Save predictions to a CSV file
        results_df.to_csv('ensemble_test_results.csv', index=False)
        print("Ensemble predictions have been saved to 'ensemble_test_results.csv'")
    except Exception as e:
        print(f"Error during ensemble prediction: {e}")
else:
    print("No predictions were made. Please check the errors in the prediction loop.")

# Additional Validation
if not all_predictions:
    print("Debugging individual models:")
    for trainer, tokenizer in trained_models:
        model_name = trainer.model.name_or_path
        print(f"Validating model: {model_name}")
        try:
            trainer.predict(test_dataset)  # Test compatibility
            print(f"Model {model_name} is compatible with the test dataset.")
        except Exception as e:
            print(f"Incompatibility with model {model_name}: {e}")
```
